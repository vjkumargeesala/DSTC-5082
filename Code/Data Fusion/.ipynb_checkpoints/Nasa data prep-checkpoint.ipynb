```python
import pandas as pd
```

```python
df_1_10=pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/POWER_Point_Hourly_20010101_20101231_032d90N_097d04W_LST.csv",skiprows=15)
df_1_10
```

```python
df_11_20=pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/POWER_Point_Hourly_20110101_20201231_032d90N_097d04W_LST.csv",skiprows=15)
df_21_24=pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/POWER_Point_Hourly_20210101_20241231_032d90N_097d04W_LST.csv",skiprows=15)
```

```python
df = pd.concat([df_1_10,df_11_20,df_21_24], ignore_index=True)
```

```python
df
```

```python
df['datetime'] = pd.to_datetime(df[['YEAR', 'MO', 'DY', 'HR']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d-%H')

df = df.drop(columns=['YEAR', 'MO', 'DY', 'HR'])
df
```

```python
df.rename(columns={
    'datetime': 'datetime',
    'T2M': 'temp_C',
    'PS': 'pressure_kPa',
    'RH2M': 'humidity',
    'WS2M': 'wind_speed',
    'PRECTOTCORR': 'rain_1h',
    'ALLSKY_SFC_LW_DWN': 'Longwave_Irradiance_Wm2',
    'ALLSKY_SFC_SW_DWN': 'Shortwave_Irradiance_Wm2'
}, inplace=True)

desired_order = [
    'datetime', 'temp_C', 'feels_like_C', 'dew_point_C',
    'temp_min_C', 'temp_max_C', 'pressure_kPa', 'humidity',
    'wind_speed', 'wind_gust', 'rain_1h', 'snow_1h',
    'clouds_all', 'visibility', 'Longwave_Irradiance_Wm2', 'Shortwave_Irradiance_Wm2'
]

df = df[[col for col in desired_order if col in df.columns]]
```

```python
df_o=pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/openweather_data_prep.csv",index_col=0)
df_n=pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/nasa_daily_data.csv",index_col=0)
```

```python
df_o['date'] = pd.to_datetime(df_o['date'])
df_n['datetime'] = pd.to_datetime(df_n['datetime'])
```

```python
df_o.rename(columns={'date': 'datetime'}, inplace=True)


df_merged = pd.merge(df_o, df_n, on='datetime', how='inner')
```

```python
df_merged
```

```python
df_merged.to_csv("databpam.csv")
```

```python
df_merged['datetime'] = pd.to_datetime(df_merged['datetime'])  # ensure datetime column
df_merged.set_index('datetime', inplace=True)
```

```python
df_merged.columns
```

```python
df_median = pd.DataFrame()
```

```python
df_median['pressure_kPa'] = df_merged[['pressure_kPa_x', 'pressure_kPa_y']].median(axis=1)
df_median['humidity'] = df_merged[['humidity_x', 'humidity_y']].median(axis=1)
df_median['wind_speed'] = df_merged[['wind_speed_x', 'wind_speed_y']].median(axis=1)
df_median['rain_1h'] = df_merged[['rain_1h_x', 'rain_1h_y']].median(axis=1)
df_median['Temp']=df_merged[['median_temp','temp_C']].median(axis=1)
```

```python
df_median.index = df_merged.index
```

```python
df_median
```

```python
df_mean=pd.DataFrame()
```

```python
df_mean['pressure_kPa'] = df_merged[['pressure_kPa_x', 'pressure_kPa_y']].mean(axis=1)
df_mean['humidity'] = df_merged[['humidity_x', 'humidity_y']].mean(axis=1)
df_mean['wind_speed'] = df_merged[['wind_speed_x', 'wind_speed_y']].mean(axis=1)
df_mean['rain_1h'] = df_merged[['rain_1h_x', 'rain_1h_y']].mean(axis=1)
df_mean['Temp']=df_merged[['median_temp','temp_C']].mean(axis=1)
df_mean.index = df_merged.index
```

```python
df_merged['pressure_kPa'] = df_merged[['pressure_kPa_x', 'pressure_kPa_y']].mean(axis=1)
df_merged['humidity'] = df_merged[['humidity_x', 'humidity_y']].mean(axis=1)
df_merged['wind_speed'] = df_merged[['wind_speed_x', 'wind_speed_y']].mean(axis=1)
df_merged['rain_1h'] = df_merged[['rain_1h_x', 'rain_1h_y']].mean(axis=1)
df_merged['Temp']=df_merged[['median_temp','temp_C']].mean(axis=1)
```

```python
df_merged.drop(
    ['pressure_kPa_x', 'pressure_kPa_y', 'humidity_x', 'humidity_y',
     'wind_speed_x', 'wind_speed_y', 'rain_1h_x', 'rain_1h_y',
     'median_temp', 'temp_C'],
    axis=1,
    inplace=True
)
```

```python
import pandas as pd

df = pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/DFW_weather.csv")


df
```

```python
df.info()
```

```python
df['dt_iso'] = df['dt_iso'].str.split('+').str[0].str.strip()
print(df.head(5))
df['dt_iso'] = pd.to_datetime(df['dt_iso'])

df = df[(df['dt_iso'].dt.year > 2000) & (df['dt_iso'].dt.year <= 2024)].copy()
```

```python
df = df.drop_duplicates()
df
```

```python
df["dt_iso"].nunique()
```

```python
df = df.drop_duplicates(subset='dt_iso', keep='first')
df
```

```python
temp_cols = ['temp', 'feels_like', 'dew_point', 'temp_min', 'temp_max']
for col in temp_cols:
    if col in df.columns:
        df[col] = df[col] - 273.15  # overwrite in place
        df.rename(columns={col: col + '_C'}, inplace=True)
df
```

```python
pressure_cols = ['pressure', 'sea_level', 'grnd_level']
for col in pressure_cols:
    if col in df.columns:
        df[col] = df[col] / 10  # overwrite in place
        df.rename(columns={col: col + '_kPa'}, inplace=True)
```

```python
df = df.rename(columns={'dt_iso': 'date'})
```

```python
df
```

```python
final_columns = [
    'date','temp_C', 'feels_like_C', 'dew_point_C',
    'temp_min_C', 'temp_max_C', 'pressure_kPa', 'humidity', 'wind_speed', 'wind_gust',
    'rain_1h', 'snow_1h', 'clouds_all', 'visibility',
    
]

df_cleaned = df[final_columns]
```

```python
df_cleaned
```

```python
df_cleaned.to_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/openweather_daily_data.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_csv("/Users/vj/Jupyter Notebook/DSTC-5082/Data/openweather_daily_data.csv")
```

```python
df
```

```python
df['avg_temp'] = df[['temp_C', 'feels_like_C', 'temp_min_C', 'temp_max_C']].mean(axis=1)
df['median_temp'] = df[['temp_C', 'feels_like_C', 'temp_min_C', 'temp_max_C']].median(axis=1)
```

```python
df
```

```python
df_melted = df.melt(id_vars='date',
                    value_vars=['temp_C', 'feels_like_C', 'temp_min_C', 'temp_max_C'],
                    var_name='Temperature_Type',
                    value_name='Temperature')
```

```python
# Plotting
plt.figure(figsize=(14, 6))
sns.set_style("whitegrid")

# Lineplot for the 4 temperature types
sns.lineplot(data=df_melted, x='date', y='Temperature', hue='Temperature_Type', linewidth=2)

# Add avg and median lines manually
#plt.plot(df['datetime'], df['avg_temp'], label='Avg Temp (Mean)', color='black', linestyle='--', linewidth=2)
#plt.plot(df['datetime'], df['median_temp'], label='Median Temp', color='gray', linestyle=':', linewidth=2)
```

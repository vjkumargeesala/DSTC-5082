```python
import pandas as pd
import dask.dataframe as dd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('/content/DFW_weather.csv')
print(df.head())
```

```python

df = df.iloc[200000:]
df['dt_iso'] = df['dt_iso'].str.replace(' UTC', '', regex=False)  # Remove ' UTC'
df['dt_iso'] = pd.to_datetime(df['dt_iso'], format='%Y-%m-%d %H:%M:%S %z', utc=True)
df.set_index('dt_iso', inplace=True)
# Convert temperature from Kelvin to Celsius
df['temp'] = df['temp'] - 273.15
```

```python
df.columns
```

```python
temp = df['temp']
temp.plot()
```

```python
import numpy as np

def df_to_X_y(df, target_col, window_size):
    feature_cols = ['temp', 'dew_point', 'wind_speed', 'pressure', 'humidity']

    df_as_np = df[feature_cols].to_numpy()
    target_np = df[target_col].to_numpy()

    X, y = [], []
    for i in range(len(df_as_np) - window_size):
        X.append(df_as_np[i:(i + window_size)])
        y.append(target_np[i + window_size])

    return np.array(X), np.array(y)


```

```python
window_size = 24
X, y = df_to_X_y(df, 'temp', window_size)
X.shape, y.shape
```

```python
X_train,y_train = X[:200000],y[:200000]
X_test,y_test = X[200000:250000],y[200000:250000]
X_valid,y_valid = X[250000:],y[250000:]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((window_size , 5)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()


```

```python
model1.compile(loss=MeanSquaredError(), metrics=[RootMeanSquaredError()], optimizer=Adam(learning_rate=0.0001))
model1.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=10)

model1.save('/content/model1.keras')
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('/content/model1.keras')
```

```python
train_predictions = model1.predict(X_train).flatten()
train_results = pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals':y_train})
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(train_results['Train Predictions'][:1000], label='Train Predictions')
plt.plot(train_results['Actuals'][:1000], label='Actuals')
plt.legend()
plt.show()
```

```python
test_predictions = model1.predict(X_test).flatten()
test_results = pd.DataFrame(data={'Test Predictions':test_predictions, 'Actuals':y_test})
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(test_results['Test Predictions'][:1000], label='Test Predictions')
plt.plot(test_results['Actuals'][:1000], label='Actuals')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Evaluate on test data
mse = mean_squared_error(y_test, test_predictions)
mae = mean_absolute_error(y_test, test_predictions)
r2 = r2_score(y_test, test_predictions)

print(f"Test MSE: {mse:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test R²: {r2:.4f}")
```

```python
# Start with the last known 24-hour window
input_seq = X_test[-1]  # Shape: (24, 5)
future_preds = []

for _ in range(24):  # Predict next 24 hours
    pred = model1.predict(input_seq.reshape(1, 24, 5), verbose=0).flatten()[0]
    future_preds.append(pred)

    # Shift the window and replace only the temperature in the last row
    next_input = np.roll(input_seq, -1, axis=0)
    next_input[-1, 0] = pred  # Replace temperature only
    input_seq = next_input

# Get actual temperatures for comparison
actual_temps = y_test[-24:]

# Plot future predictions vs actuals
plt.figure(figsize=(10, 5))
plt.plot(future_preds, marker='o', label='24-hour Forecasted Temps')
plt.plot(actual_temps, marker='x', label='Actual Temps')
plt.title('Recursive 24-hour Temperature Forecast vs Actual')
plt.xlabel('Hour Ahead')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def df_to_X_y(df, target_col, window_size):
    feature_cols = ['temp', 'dew_point', 'wind_speed', 'pressure', 'humidity']

    df_as_np = df[feature_cols].to_numpy()
    target_np = df[target_col].to_numpy()

    X, y = [], []
    for i in range(len(df_as_np) - window_size):
        X.append(df_as_np[i:(i + window_size)])
        y.append(target_np[i + window_size])

    return np.array(X), np.array(y)
```

```python
target_cols = [ 'humidity']
window_size = 24
X, y = df_to_X_y(df, target_cols, window_size)

X_train, y_train = X[:200000], y[:200000]
X_test, y_test = X[200000:250000], y[200000:250000]
X_valid, y_valid = X[250000:], y[250000:]
```

```python
model1 = Sequential()
model1.add(InputLayer((window_size, 5)))
model1.add(LSTM(128, return_sequences=True))
model1.add(Dropout(0.2))
model1.add(LSTM(64))
model1.add(Dense(32, activation='relu'))
model1.add(Dense(1, activation='linear'))

model1.compile(
    loss=MeanSquaredError(),
    optimizer=Adam(learning_rate=0.001),
    metrics=[RootMeanSquaredError(), 'mae']
)

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_schedule = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.5)

model1.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    epochs=10,
    callbacks=[early_stop, lr_schedule]
)
model1.save('/content/modelhumidity.keras')
```
